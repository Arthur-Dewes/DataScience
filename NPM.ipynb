```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000) #gera 1000 pontos no intervalo [-5, 5]

sigmoide = lambda x: 1 / (1 + np.exp(-x))#sigmoide eq
relu = lambda x: np.maximum(0, x)#ReLU eq
tanh = lambda x: (np.e**x - (1/np.e**x)) / (np.e**x + (1/np.e**x))#hyperbolic tangent eq
gelu =  lambda x: x * (1 + tanh(np.sqrt(2/np.pi) * (x + 0.044715*x**3)))/ 2#GeLu eq
leaky_relu = lambda x: max(0.01*x,x)#Leaky ReLU eq
swish = lambda x: x / (1 + np.exp(-x))#Swish eq
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3)

ax1.plot(x, np.array(list(map(sigmoide,x)),dtype = 'float16'), label="Sigmoid", color="blue")
ax1.grid()
ax1.legend(fontsize="6",loc='upper left')

ax2.plot(x, np.array(list(map(relu,x)),dtype = 'float16'), label="ReLU", color="red")
ax2.grid()
ax2.legend(fontsize="6",loc='upper left')

ax3.plot(x, np.array(list(map(leaky_relu,x)),dtype = 'float16'), label="Leaky ReLU", color="green")
ax3.grid()
ax3.legend(fontsize="6",loc='upper left')

ax4.plot(x, np.array(list(map(tanh,x)),dtype = 'float16'), label="Tanh", color="olive")
ax4.grid()
ax4.legend(fontsize="6",loc='upper left')

ax5.plot(x, np.array(list(map(gelu,x)),dtype = 'float16'), label="GeLU", color="magenta")
ax5.grid()
ax5.legend(fontsize="6",loc='upper left')

ax6.plot(x, np.array(list(map(swish,x)),dtype = 'float16'), label="Swish", color="cyan")
ax6.grid()
ax6.legend(fontsize="6",loc='upper left')

plt.show()
```
